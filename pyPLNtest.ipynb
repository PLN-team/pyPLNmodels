````python
import pandas as pd 

Y = pd.read_csv('example_data/Y_test')
O = pd.read_csv('example_data/O_test')
cov = pd.read_csv('example_data/cov_test')
````

###  Fit a PLNmodel 

````python
from pyPLNmodels.models import PLNmodel

q = 5
mypln = PLNmodel(q)
mypln.fit(Y,O,cov)
print(mypln)
````

### PLNmodel maximizing log likelihood

````python
nbpcs = 5 # number of principal components 
mypln = PLNmodel(q= nbpcs)
mypln.fit(Y,O,cov, fast = False)
print(mypln)
````

### Get the parameters

````python
beta = mypln.get_beta()
C = mypln.get_C()
Sigma = mypln.get_Sigma()
````

## How to fit each model? 

The object PLNmodel calls one model depending on the value of ```q``` and ```fast```. It chooses among 3 models. You can fit any of them if you prefer to use one model among another. Basically, fastPLN is fast when ```p``` is low ( $\leq 8000$). fastPLNPCA is faster for ```p```$>8000$. IMPS_PLN is always slower and uses a more complicated method to fit the model but is more 

### Fit the PLN model

````python
from pyPLNmodels.models import fastPLN
fast = fastPLN()
fast.fit(Y,O,cov)
print(fast)
````

### Fit the fastPLNPCA model

````python
from pyPLNmodels.models import fastPLNPCA
nbpcs = 5
fastpca = fastPLNPCA(q=nbpcs)
fastpca.fit(Y,O,cov)
print(fastpca)
````

### Fit the IMPS_PLN model

````python
from pyPLNmodels.models import IMPS_PLN
nbpcs = 5
imps = IMPS_PLN(nbpcs)
imps.fit(Y,O,cov)
print(imps)
````
